**Resizing image to common dimension**

**Converting RGB to Grey Scale images using OpenCV**

```python
#importing openCV library
import cv2
```

```python
img=cv2.imread('/content/beagle-hound-dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
type(greyscale_img)
```

```python
greyscale_img.shape
```

cv2.imshow(greyscale_img) will display the image, but this is not allowed in Google Colab

we use

**from google.colab.patches import cv2_imshow**

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(greyscale_img)
```

```python
#saving grey scale image from cv2
cv2.imwrite('dog_grayscale_image.jpg',greyscale_img)
```
